```python
from dbfread import DBF
import pandas as pd
from pandas import DataFrame
from itertools import chain
```

```python
# create a df for the first day
day=20140701
dbf = DBF(r"C:\Users\taohuang\Downloads\prism-main20221226\prism-main\output\PRISM_vpdmax_stable_4kmD2_"+str(20140701)+".dbf")
frame = DataFrame(iter(dbf))
frame['date_num']=day
frame_all=frame
frame_all
```

```python
res = chain(range(20140701,20140731+1),
            range(20140801,20140829+1) )

for day in res:
    dbf = DBF(r"C:\Users\taohuang\Downloads\prism-main20221226\prism-main\output\PRISM_vpdmax_stable_4kmD2_"+str(day)+".dbf")
#        dbf = DBF(r"C:\Users\taohuang\Downloads\prism-main\PRISM_tmax_stable_4kmD2_"+str(20210830)+".dbf")
    frame = DataFrame(iter(dbf))
    frame['date_num']=day
    frame_all=frame_all.append(frame)
    
frame_all['date'] = pd.to_datetime(frame_all['date_num'], format='%Y%m%d')
    
print(day,frame_all)
```
